```python
df1= df.loc[df['kills'] > 0]
```

```python
df1
```

```python
df2 = df1.sort_values(by='kills', ascending= False).reset_index()
df2
```

```python
df3 = df2.groupby(["Id"])['kills'].max().reset_index()
df3.columns = ["Id","kills"]
df3
```

```python
#select the style of the plot 
style.use('ggplot')
#plot the random number 
plt.plot(df3['kills'],'r', label = 'kills', linewidth = 1) #here 'line one' is the lagend
#x axis is the number of random numbers (index)
plt.xlabel('Total User')
#y axis is actual random number 
plt.ylabel('Total Kills')
#title
plt.title('Plot data as total kills by user')

plt.legend()
plt.show()
```

```python
df4= df.loc[df['kills'] <= 0]
df4
```

```python
df4.columns
```

```python
subset = df4.loc[:, ['Id','kills','damageDealt','winPlacePerc']]
```

```python
#select the style of the plot 
style.use('ggplot')
#plot the random number 
plt.plot(subset['damageDealt'],'g', label = 'Total Damage', linewidth = 1) #here 'line one' is the lagend
#x axis is the number of random numbers (index)
plt.xlabel('Total User')
#y axis is actual random number 
plt.ylabel('Total Damage by user')
#title
plt.title('Plot data as total damage by user who has the zero kills')

plt.legend()
plt.show()
```
